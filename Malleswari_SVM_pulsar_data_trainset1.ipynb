#### I am imputing the median of Excess kurtosis of the integrated profile and Standard deviation of the DM-SNR curve  because Excess kurtosis of the integrated profile has mean and median are near 

```python
pulsar_data_train[' Excess kurtosis of the integrated profile']=pulsar_data_train[' Excess kurtosis of the integrated profile'].fillna(
    pulsar_data_train[' Excess kurtosis of the integrated profile'].median())
```

```python
pulsar_data_train[' Standard deviation of the DM-SNR curve']=pulsar_data_train[' Standard deviation of the DM-SNR curve'].fillna(
    pulsar_data_train[' Standard deviation of the DM-SNR curve'].median())
```

#### Here there are some negative corelation is there so i can check the VIF for more clarity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = pd.DataFrame()
vif_data["feature"] = pulsar_data_train.columns
vif_data["VIF"] = [variance_inflation_factor(pulsar_data_train.values, i)
                          for i in range(len(pulsar_data_train.columns))]
  
print(vif_data)
```

### By observing the above cells Excess kurtosis of the DM-SNR curve and Mean of the integrated profile has more variance and correlation to the target variable is also negative so i delete both columns

```python
pulsar_data_train=pulsar_data_train.drop([' Mean of the integrated profile',' Excess kurtosis of the DM-SNR curve'],axis=1)
```

```python
vif_data = pd.DataFrame()
vif_data["feature"] = pulsar_data_train.columns
vif_data["VIF"] = [variance_inflation_factor(pulsar_data_train.values, i)
                          for i in range(len(pulsar_data_train.columns))]
  
print(vif_data)
```
